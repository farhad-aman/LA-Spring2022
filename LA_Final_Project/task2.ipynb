```python
import matplotlib.pyplot as plt
import numpy as np 
from PIL import Image
plt.rcParams["figure.figsize"] = (12, 8.25)
```

```python
image = Image.open('Source4.jpg')
plt.imshow(image)
source = np.asarray(image, dtype=float)
source.shape
```

```python
image = Image.open('Reference4.jpg')
plt.imshow(image)
reference = np.asarray(image, dtype=float)
reference_pixels = reference.shape[0] * reference.shape[1]
reference_pixels
```

```python
#Red
#Find mean of red in source and reference

reference_red = np.sum(reference[:, :, 0]) / (reference.shape[0] * reference.shape[1])
print(reference_red)
source_red = np.sum(source[:, :, 0]) / (source.shape[0] * source.shape[1])
print(source_red)
diff_red = reference_red - source_red
for i in range(source.shape[0]):
    for j in range(source.shape[1]):
        if source[i][j][0] + diff_red >= 0 and source[i][j][0] + diff_red < 256:
            source[i][j][0] += diff_red
        elif source[i][j][0] + diff_red < 0:
            source[i][j][0] = 0
        else:
            source[i][j][0] = 255
```

```python
#Green
#Find mean of green in source and reference

reference_green = np.sum(reference[:, :, 1]) / (reference.shape[0] * reference.shape[1])
print(reference_green)
source_green = np.sum(source[:, :, 1]) / (source.shape[0] * source.shape[1])
print(source_green)
diff_green = reference_green - source_green
for i in range(source.shape[0]):
    for j in range(source.shape[1]):
        if source[i][j][1] + diff_green >= 0 and source[i][j][1] + diff_green < 256:
            source[i][j][1] += diff_green
        elif source[i][j][1] + diff_green < 0:
            source[i][j][1] = 0
        else:
            source[i][j][1] = 255
```
